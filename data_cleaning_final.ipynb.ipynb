```python
from datasets import load_dataset

ds = load_dataset("0x70DA/drone-spectrogram")
print(ds)
```

```python
import os
from PIL import Image

os.makedirs("data/drone", exist_ok=True)

for i, item in enumerate(ds["train"]):
    img = item["image"]
    img.save(f"data/drone/img_{i}.png")

print("Done saving images")
```

```python
from PIL import Image
import os

folder = "data/drone"

for file in os.listdir(folder):
    path = os.path.join(folder, file)
    try:
        img = Image.open(path)
        img = img.resize((128,128))
        img.save(path)
    except:
        os.remove(path)

print("Cleaning done")
```
